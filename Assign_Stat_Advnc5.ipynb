```python
# Q-1
from scipy.stats import norm

# Sample mean
sample_mean = 50
# Sample standard deviation
sample_std = 5
# Desired confidence level
confidence_level = 0.95

# Calculate the confidence interval
lower_bound, upper_bound = norm.interval(confidence_level, loc=sample_mean, scale=sample_std)

print("95% Confidence Interval:", (lower_bound, upper_bound))
```

```python
# Q-2
from scipy.stats import chisquare

# Observed frequencies (actual counts of each color)
observed_freq = [10, 25, 15, 5, 20, 25]  # Blue, Orange, Green, Yellow, Red, Brown

# Expected frequencies based on the expected distribution
expected_freq = [0.20*100, 0.20*100, 0.20*100, 0.10*100, 0.10*100, 0.20*100]  # 100 M&Ms in total

# Perform the chi-square goodness of fit test
chi2_stat, p_value = chisquare(observed_freq, f_exp=expected_freq)

# Print the results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

# Check if the null hypothesis is rejected or not
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The distribution of colors of M&Ms in the bag does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of colors of M&Ms in the bag matches the expected distribution.")
```

```python
# Q-3
from scipy.stats import chi2_contingency

# Define the contingency table
contingency_table = [[20, 15],
                     [10, 25],
                     [15, 20]]

# Perform the chi-square test of independence
chi2_stat, p_value, dof, expected_freq = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is significant evidence of association between the two categorical variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant evidence of association between the two categorical variables.")
```

```python
# Q-4
import scipy.stats as stats

# Sample size
n=500
# Number of individuals who smoke in the sample
x=60
# Proportion of individuals who smoke in the sample
p=x/n

# Calculate the standard error
standard_error = stats.norm.ppf(0.975)*((p*(1-p)) / n)**0.5

# Calculate the margin of error
margin_of_error = standard_error

# Calculate the confidence interval
lower_bound = p - margin_of_error
upper_bound = p + margin_of_error

print("95% Confidence Interval for the true proportion of individuals who smoke is:", (lower_bound, upper_bound))
```

```python
# Q-6
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the degrees of freedom
df = 10

# Generate x values for the chi-square distribution
x = np.linspace(0, 30, 1000)

# Plot the chi-square distribution
plt.plot(x, stats.chi2.pdf(x, df), 'b-', label=f'Chi-square distribution (df={df})')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(15, 30, 1000)
plt.fill_between(x_fill, stats.chi2.pdf(x_fill, df), color='orange', alpha=0.5, label='Chi-square statistic of 15')

# Add labels and title
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square Dist')
```

```python
# Q-7
import scipy.stats as stats

# Sample size
n = 1000
# Number of people who preferred Coke in the sample
x = 520
# Proportion of people who preferred Coke in the sample
p_hat = x / n

# Calculate the standard error
standard_error = (p_hat * (1 - p_hat) / n) ** 0.5

# Calculate the margin of error using the z-score for a 99% confidence interval
z_score = stats.norm.ppf(0.995)  # 99% confidence level for a two-tailed test
margin_of_error = z_score * standard_error

# Calculate the confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print("99% Confidence Interval for the true proportion of people who prefer Coke:", (lower_bound, upper_bound))
```

```python
# Q-11
import scipy.stats as stats

# Sample information
sample_mean = 72
sample_std = 10
sample_size = 30

# Population mean under the null hypothesis
population_mean_null = 70

# Calculate the t-score
t_score = (sample_mean - population_mean_null) / (sample_std / (sample_size ** 0.5))

# Calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_score), df=sample_size - 1))

# Print the results
print("t-score:", t_score)
print("p-value:", p_value)

# Check if the null hypothesis is rejected or not
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: There is no significant evidence to conclude that the population mean is different from 70.")
```
